```python
# Top 10 categorías
top_categorias = df.groupby('category')['engagement_rate'].mean().sort_values(ascending=False).head(10).reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=top_categorias, y='category', x='engagement_rate', palette='viridis')
plt.title('Top 10 Categorías con Mayor Engagement Rate', fontsize=14, fontweight='bold')
plt.xlabel('Engagement Rate')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
# Agrupar por hora
hora_engagement = df.groupby('upload_hour')['engagement_rate'].mean().reset_index()

plt.figure(figsize=(12, 5))
sns.lineplot(data=hora_engagement, x='upload_hour', y='engagement_rate', marker='o', color='#FF6600', linewidth=2)
plt.title('Engagement Rate según Hora de Publicación', fontsize=14, fontweight='bold')
plt.xlabel('Hora del día (0-23)')
plt.ylabel('Engagement Rate')
plt.xticks(range(0, 24))
plt.grid(True, alpha=0.3)
plt.axhline(y=hora_engagement['engagement_rate'].mean(), color='gray', linestyle='--', label='Promedio general')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Datos
emoji_engagement = df.groupby('has_emoji')['engagement_rate'].mean().reset_index()
emoji_engagement['has_emoji'] = emoji_engagement['has_emoji'].map({0: 'Sin emoji', 1: 'Con emoji'})

plt.figure(figsize=(6, 5))
colors = ['#A0A0A0', '#FFD700']
sns.barplot(data=emoji_engagement, x='has_emoji', y='engagement_rate', palette=colors)
plt.title('Impacto de los Emojis en el Engagement', fontsize=14, fontweight='bold')
plt.ylabel('Engagement Rate')
plt.xlabel('')

# Etiquetas
for i, row in emoji_engagement.iterrows():
    plt.text(i, row['engagement_rate'] + 0.002, f"+{((row['engagement_rate']/0.07023)-1)*100:.0f}%", ha='center', fontweight='bold')

plt.ylim(0, 0.1)
plt.show()
```

```python
# Calcular percentiles
percentil_75 = df['engagement_rate'].quantile(0.75)
percentil_90 = df['engagement_rate'].quantile(0.90)

plt.figure(figsize=(10, 5))

# Gráfico simplificado: boxplot con puntos destacados
sns.boxplot(y=df['engagement_rate'], color='lightblue', width=0.3)
sns.stripplot(y=df['engagement_rate'], color='darkblue', alpha=0.3, size=2)

# Destacar percentiles
plt.axhline(y=percentil_75, color='green', linestyle='--', label=f'Top 25%: {percentil_75:.1%}')
plt.axhline(y=percentil_90, color='orange', linestyle='--', label=f'Top 10%: {percentil_90:.1%}')
plt.axhline(y=df['engagement_rate'].mean(), color='red', linestyle='-', label=f'Promedio: {df["engagement_rate"].mean():.1%}')

plt.title('Distribución del Engagement Rate: ¿Dónde está el éxito?', fontsize=14, fontweight='bold')
plt.ylabel('Engagement Rate')
plt.legend()
plt.tight_layout()
plt.show()

# Imprimir resumen ejecutivo en texto
print("📊 RESUMEN EJECUTIVO:")
print(f"• Promedio general: {df['engagement_rate'].mean():.1%}")
print(f"• Un video está en el top 25% si supera {percentil_75:.1%}")
print(f"• Un video está en el top 10% (viral) si supera {percentil_90:.1%}")
```
